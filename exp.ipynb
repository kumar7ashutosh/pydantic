```python
from typing import Optional,Annotated
class state(BaseModel):
    name:Annotated[str, Field(max_length=50,title="Name",description=".....")]
    age:Annotated[int,Field(gt=0,lt=100)]
    weight:Annotated[float,Field(gt=0,strict=True)]
    married:Annotated[bool,Field(default=None,description="Married or not")]
    allergies:Annotated[list[str],Field(default=None,description="Allergies")]
    contact_info:dict[str,str]
    
data=state(name="a",age=5,weight=50.0,allergies=["almonds"],contact_info={"phone":"+91xxxxxxx"})
print(data)
```

```python
import re
class state(BaseModel):
    name:str
    age:int
    weight:float
    married:Optional[bool]=None
    allergies:list[str]
    contact_info:dict[str,str]
    @field_validator('contact_info')
    @classmethod
    def check_contact_info(cls, value: dict[str, str]):
        if "phone" not in value:
            raise ValueError("contact_info must contain a phone number")

        phone = value["phone"]

        if not re.fullmatch(r"\+\d{7,15}", phone):
            raise ValueError(
                "phone number must be in format +<country_code><number>"
            )

        return value
            
data=state(name="a",age=5,weight=50.0,allergies=["almonds"],contact_info={"phone":"+919876543210"})
print(data)
```

```python
import re
class state(BaseModel):
    name:str
    age:int
    weight:float
    married:Optional[bool]=None
    allergies:list[str]
    contact_info:dict[str,str]
    @field_validator('contact_info')
    @classmethod
    def check_contact_info(cls, value: dict[str, str]):
        if "phone" not in value:
            raise ValueError("contact_info must contain a phone number")

        phone = value["phone"]

        if not re.fullmatch(r"\+\d{7,15}", phone):
            raise ValueError(
                "phone number must be in format +<country_code><number>"
            )

        return value
    @model_validator(mode="after")
    def validate_emergency_contact(cls,model):
        if model.age>60 and model.contact_info.get("emergency_contact") is None:
            raise ValueError("Emergency contact is required for age > 60")
            return model
data=state(name="a",age=5,weight=50.0,allergies=["almonds"],contact_info={"phone":"+919876543210"})
print(data)
```
